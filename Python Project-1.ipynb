```python
df2 = df.dropna()
df2 = df2[df2['Number Scoring 80 or Above'] != 's']
```

# Selecting my school and analysis

---

```python
df3 = df2[(df2['School Name'] == 'Abraham Lincoln High School')]
```

```python
df3
```

```python
## calcutating mean based on year
table = pd.pivot_table(df3, values='Number Scoring 80 or Above', index=['Year'], columns=['School Name'], aggfunc=np.mean, fill_value=0)
```

```python
table
```

```python
df3.groupby(['Year'])['Number Scoring 80 or Above'].count()
```

```python
df3.groupby(['Year'])['Number Scoring 80 or Above'].count().plot(kind='bar')
```
